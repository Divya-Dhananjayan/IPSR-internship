Step 2: Data Merging
=====================


First we are trying to merge the first 2 data frames df1 and df2.Then the resultant is merged with the third data frame df3 . Thus we merge whole data into a single one


For merging we are not considering the whole columns of each data sets.
We are picking some of the required columns only

```python

# From df1 we are excluding state and city columns as these columns are not cleaned


df1 = df1[[
    'ranking_year','ranking_category',
    'institute_id','institute_name',
    'category','no. of phd students'
    ]]
```

```python

# From df2 we are excluding state and city columns as these columns are not cleaned

df2 = df2[[
    'ranking_year','ranking_category',
    'institute_id','institute_name',
    'score','rank'
     ]]

```

```python
# 'Overa all score' column is renamed to score as it is same as the score column of df3


df3_splitted.rename(columns= {'Overall Score': 'score'} ,inplace = True)
```

```python
# From df3_splitted we are excluding state and city columns as these columns
# are not cleaned
# The resultant is saved to df3 for convenience

df3 = df3_splitted[
    ['ranking_year','ranking_category','institute_id','institute_name',
     'Graduation Outcomes','Outreach and Inclusivity','score','Perception','Qualitative Research'
     ,'Quantitative Research','Research and Professional Practice','Students and Faculty Contribution',
     'Teaching, Learning and Resources']
    ]
```

```python
# Merging df1 and df2


df_merged_df1df2 = pd.merge(left = df1 , right = df2 ,
                            on = ['ranking_year','ranking_category',
                                  'institute_id','institute_name']
                                  ,how = 'outer')

```

```python

# The resultant is trying to merge with df3

df_merged_df1df2df3 = pd.merge(left =df_merged_df1df2 , right = df3 ,
                               on=['ranking_year','ranking_category',
                                   'institute_id','institute_name','score']
                               ,how = 'outer')
```
