```python
# Extra spaces and special characters from the institute_name column is removed


df1['institute_name'] = df1['institute_name'].str.lower().str.replace(r'[^a-z0-9 ]', '', regex=True).str.strip()


# trying to make one to one institute_id and institute_name
# keeping the first occurance of institute name for the institute id

grpd = df1.groupby('institute_id')['institute_name'].first()


# replacing the institute name column with unique name of institute name

df1['institute_name']= df1['institute_id'].map(grpd)


#  Now  chances for duplicate entries are there


df1.drop_duplicates()
```

See the number of entries decrease to 8916 from 9342

let's check the unique number of institute_id and institute_name in df1

```python
df1[['institute_id','institute_name']].nunique()
```

The difference in their number is beacuse many institutions are multidisciplinary institutes having same institute_id for all their departments

```python
#Extracting year from 'academic_year' column

df1['academic_year'] =df1['academic_year'].str.extract(r'(\d{4}[-_]\d{2})')
df1['academic_year']


```

```python

# Renaming the 'value' column

df1.rename(columns= {'value':'no. of phd students'} ,inplace= True)

```

```python
# Removing the unit column

df1.drop('unit',axis=1,inplace = True)

```

```python
# checking number of unique institute id in df2

df2['institute_id'].nunique()
```

In the second file having score and rank of institutes have data from 2016 onwards. And the category field have additional 5 entries compared to first file . Here the data is more

we know there are complicated issues in the institute name column , so my plan is to replace the institute name of df2 with institute name of df1 according to the institute id.Both the files have institute id same .

```python
# Making a data frame having  the 'institute_id' and the 'institute_name' columns
#columns from df1

df1_institute_id_name = df1[['institute_id','institute_name']]

# AS we  are considering only two parameters id and name our new data frame
# have the same rows multiple times. So we must remove the duplicate entries

df1_institute_id_name = df1_institute_id_name.drop_duplicates(subset= 'institute_id')

# restoring the result to the same dataframe

df1_institute_id_name
```


Now we got it right. But we are trying to map this to df2. Map function works only on series. Not with dataframe so we convert it to series.



```python

# creating series having institute id as index and column as institute-name

id_correspondingto_name = df1_institute_id_name.set_index('institute_id')['institute_name']
id_correspondingto_name

```

```python


# Now replacing the institute name column of df2
#filling institute name considering the institute id and filling accordingly with
# the corresponding institute name from df1



df2['institute_name'] = df2['institute_id'].map(id_correspondingto_name).fillna(df2['institute_name'])

```

```python

# dropping duplicates

df2.drop_duplicates()
```

Now we got a cleaned institute name column


Now consider the score column

```python

# handling missing values in score column

df2[df2['score'].isna()]
```

while analysing the score column I identified that for the ranking category 'Innovation','Skill University','Open University' they dont have any score data even in the original data from the website. So we can consider it separately.

We can keep the missing data as missing data itself.

Considering the rank column

df2["rank"].astype(int)

 This line of code is used to change the column data type, but it is showing error. From this it is understood that it has some non integer values. we need to find out that.

Trying to figure out the nonint values in the rank column




```python
# Trying to convert 'rank' column to numeric . As it is not directly possible
# (we have seen before witth df2["rank"].astype(int) ) so convert  those values to nan

convert_to_numeric = pd.to_numeric(df2['rank'], errors = 'coerce')

# a boolean mask created

rows = convert_to_numeric.isna() | (convert_to_numeric % 1 !=0)

# rows will be true for non numeric and flaot valus

non_numeric_values = df2.loc[rows,'rank']

print(non_numeric_values)
```

We need to work on this 9 values

The values 26A,21A and so on means there are multiple institues have same rank so keeping one as 21 and the coming ones as 21A, 21B like that the pattern is following. So we can ignore the Suffix

```python
# removing A for converting rank column to int type

df2['rank']=df2['rank'].str.replace(r'A$',"",regex = True)

```

```python
# Drop the row having Rank withdrawn
df2 = df2[df2['rank'] != 'Rank Withdrawn']
```

We must study about this institution how they lost their rank, It may help our project

```python
df2['rank'].dtypes
```

```python
# we cannt convert a float type or non numeric object datatype directly to int
# here df2['rank'].astype(int) raising error

df2["rank"] = pd.to_numeric(df2["rank"], errors="coerce").astype("Int64")

```

```python
# Now checking the data type

df2['rank'].dtypes
```

```python
# Saving the cleaned file to my local system

df2.to_csv('NIRF_File2.csv')
```
