```python
import pandas as pd
```

```python
df = pd.read_csv('/content/3dfs_merged.csv')
```

```python
df
```

```python
df_engg = df[df['ranking_category']=='Engineering']
df_engg
```

```python
df_engg.to_csv('Engineering.csv')
```

```python
def top10(year):
    df = df_engg[df_engg['ranking_year'] == year].sort_values('rank')
    return df.head(10)

```

```python
def bottom10(year):
    df = df_engg[df_engg['ranking_year'] == year].sort_values('rank', ascending=False)
    return df.head(10)
```

```python
top10(2024)
```

```python
# Step 1: Filter for Part Time and Full Time rows
phd_selected = df_engg[df_engg['category'].isin(['Part Time', 'Full Time'])]

# Step 2: Group and sum PhD students
phd_total = phd_selected.groupby(['institute_name', 'ranking_year', 'rank'])['no. of phd students'].sum().reset_index()

# Step 3: Rename the column (without inplace!)
phd_total = phd_total.rename(columns={'no. of phd students': 'phd_students_total'})

df_engg_phd_summed = df_engg.merge(
    phd_total,
    on=['institute_name', 'ranking_year', 'rank'],
    how='outer')

# Step 5: Check result
print(df_engg_phd_summed.head())
```

```python
df_engg_phd_summed.to_csv('engg_phd_total.csv')
```
